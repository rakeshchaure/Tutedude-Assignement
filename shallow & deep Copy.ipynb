```python
d1 = {'id':1111,'name':'john','marks':{'eng':71.4,'maths':91.5,'bio':89.0}}
# deep copy
d2 = copy.deepcopy(d1)
d1['name']='Dan'
d1['marks']['maths']=98.9
print(f'd1-->{d1}',id(d1))
print(f'd2-->{d2}',id(d2))
```
