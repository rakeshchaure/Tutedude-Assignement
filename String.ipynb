# F-string 

```python
name="John"
age=30
language="Python"
hours=3
# without F-string
print(name,"is",age,"year old. He studies",language,hours,"hours a day.")
#with f-string
print(f"{name} is {age} year old. he studies {language} {hours} hours a day.")
```

```python
sub1=78
sub2=87
sub3=83
print(f"{name} scored {sub1 + sub2 + sub3} marks is total.")
percent=(sub1+sub2+sub3)/3
print(f'{name} and {percent}')
```

# Escape Sequence

```python
# \n - new/next line
# \t - tab
# \\ - backslash
# \' - inserts a single quotes inside a single-quoted string
##
# \n 
print("Hello everyone.\nhow are you?")

# \t 
print("John 20")
print("john\t20")

# \\
print("new\\old")
print("yes\\no")

# \'
print('this is Python\'s class')
print("This is Python's class")

# \"
print("He Says,\"we are learning Python\"")
```

```python
'''s1="Python is Fun"
print(s1[0])
print(s1[-1])
print(len(s1))

language="Python"
version ="3.13.3"
print(language + version)'''

s1="Python"
print(s1*3)

# Membership Operation
# in
s1= "Python is Fun"
print("z" in s1)
print("Java" in s1)

# not in
print("z" not in s1)
print("Java" not in s1)
print("Python" not in s1)

# Comparision of String
print("Python" == "Python")

# Removing speaces from a string -strip()

s1 ="Python"
print(s1.strip() == "Python")

# Replace()
s1= "we are learning Python"
print(s1)
print(s1.replace("Python","Java"))
print(s1)
```

# String Operation

```python
# Counting substring from a string
#count
#string.count(substring)
s1="We are learning Python. Python is fun."
s2="Python"
print(f"Occurances of space is {s1.count(s2)}")
```

```python
# Changing case of string
#upper(), lower(), title(), capitalize()
s1= "we are learning Python.Python is FUN!!."
print(s1.upper())
print(s1.lower())
print(s1.title())
print(s1.capitalize())
```

```python
#Starting and ending of a string
s1 = "We are learning Python."

#startswith()
# string.startswith(substring)

# startwith()
print(s1.startswith("We"))

# endswith()
print(s1.endswith("Python."))
```
