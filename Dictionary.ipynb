# Dictionary

```python
# Comma seperated key-value pairs enclosed within {}
#{Key1:Value1, Key2:Value2,....}

groceries ={'milk':60,'biscuits':20,'rice':90,'bread':30}
print(groceries)
print(type(groceries))
print(len(groceries))
```

# Operation on dict -->updating and deleting

```python
student1={"maths":80.5,"eng":76.0,"phy":89.0}

# fetch the marks for "phy"
print(student1["phy"])
# print(student1["chem"]) #it will be error this line beacuse not support unwanted data

#get()
print(student1.get("phy"))
print(student1.get("chem")) # it will be answere is none
print(student1.get("chem",30.40))
```

```python
emp1 ={"id":1001,"name":"john","salary":10000}
print(emp1.get("id",8959566555))

# membership operator -->
print(1001 in emp1)
print('name'in emp1)

# adding
emp1["phone"]=9993039030
print(emp1)
```

```python
sem1_marks={"Maths":78.5,"eng":71.0,"phy":86.5}
sem2_marks={"chem":81.5,"bio":90.5}

sem1_marks.update(sem2_marks)
print(sem1_marks)
```

```python
groceries_1={"milk":60,"rice":110,"biscuits":20}
groceries_2={"rice":150,"bread":30}

groceries_1.update(groceries_2)
print(groceries_1)

# pop()
groceries_1.pop("milk")
print(groceries_1)
```

```python
gro={"milk":60,"rice":100,"biscyits":20,"milk":65}
print(gro)

# keys cannot be duplicated in dict
```

# Working with keys & Values

```python
d1={(1,2,3):4}
print(d1)

# Not Allowed: list,set,dict
# allowed keys - str,int,float,bool,tuple
# keys of a dictionary can be mutable datataypes!
```

```python
# Values can be any datatype
student1={"id":1001,"name":"john","marks":[89.5,71.2,50.6,99.9]}
print(student1)
print(student1['marks'])
print(student1['marks'][1])
# fetch the keys()
# keys()
print(student1.keys(),type(student1.keys()))
# values()
print(student1.values(),type(student1.values()))
# items()
print(student1.items(),type(student1.items()))
```
