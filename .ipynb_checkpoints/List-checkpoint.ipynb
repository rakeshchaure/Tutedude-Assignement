```python
# Append
# add an item to the end of the list
fruits = ["Mango","Apple","Orange"]
print(fruits)

# syntax: list.append(item)
print(fruits.append("Banana"))
print(fruits)

s1="Python is fun"
print(s1.replace("Python","java"))


# insert
# add an element before the specified index
# syntax: list.insert (index, item)
fruits.insert(2,"graps")
print(fruits)
```

```python
# extend,remove,pop() function

# extend
fruits =["Apple","Mango","Orange"]
print(fruits)
fruits.extend(["graps","limbu"])
print(fruits)
print(len(fruits))
```

```python
# remove
fruits=["Apple","Mango","Orange"]
print(fruits)
fruits.remove("Mango")
print(fruits)
```

```python
# pop()
fruits =["Apple","Mango","Orange","Mango"]
print(fruits)
fruits.pop(-1)
print(fruits)
```

```python
"""
reverse()
sort()
count()
Membership Operation
"""
days_of_week=["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
print(days_of_week)
days_of_week.reverse()
print(days_of_week)

nums=[3,5,4,6,1,0]
print(nums)
nums.sort()
print(nums)
nums.sort(reverse=True)
print(nums)
```

```python
numbers=[10,4,-1,20,5.5,7,1]

#smallest number in the list
#min()
print(f"Smallest number:{min(numbers)}")

# Biggest number in the list
print(f"Biggest number:{max(numbers)}")
```
